```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("StudentsPerformance (1).csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
average = df[['math score','writing score','reading score']].mean()
print("Average Scores")
print(average)
```

```python
maximum = df[['math score','writing score','reading score']].max()
print("Maximum Scores")
print(maximum)
```

```python
minimum = df[['math score','writing score','reading score']].min()
print("Minimum Scores")
print(minimum)
```

```python
df["Average Score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3

print(df[["math score",
          "reading score",
          "writing score",
          "Average Score"]].head())
```

```python
highest = df.loc[df["Average Score"].idxmax()]

print("\nHighest Scoring Student")
print(highest)
```

```python
lowest = df.loc[df["Average Score"].idxmin()]

print("\nLowest Scoring Student")
print(lowest)
```

```python
gender_avg = df.groupby("gender")[[
    "math score",
    "reading score",
    "writing score"
]].mean()

print("\nGender-wise Average")
print(gender_avg)
```
