```python
import pandas as pd
```

```python
#Create a Pandas dataframe manually
data = {
    "Name":['Aryan','Rahul','Rohan'],
    "Age":[22,23,21],
    "Marks":[85,90,88]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Handle Missing values
import numpy as np

data = {
    "Name":['Aryan','Rahul','Rohan'],
    "Age":[22,np.nan,21],
    "Marks":[85,90,np.nan]
}

df = pd.DataFrame(data)

df['Age'] = df['Age'].fillna(0)
df['Marks'] = df['Marks'].fillna(0)

print(df)
```

```python
#Sort dataframe by column
data = {
    "Name":['Aryan','Rahul','Rohan'],
    "Age":[22,23,21],
    "Marks":[85,90,88]
}

sorted_df = df.sort_values(by='Marks',ascending=True)
print(sorted_df)
```

```python
#Find MEAN MEDIAN MODE 
data = {
    'Marks':[88,90,88,95,93]
}

df=pd.DataFrame(data)

print("Mean:" ,df['Marks'].mean())
print("Median:" ,df['Marks'].median())
print("Mode:" ,df['Marks'].mode()[0])
```
